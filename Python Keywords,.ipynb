```python
# Explain the significance of Python keywords and provide examples of five keywords
#if:
#Significance: Used for conditional statements, allowing me to execute a block of code if a certain condition is true.

x = 10
if x > 7:
    print("x is greater than 7")
    
#for:

#Significance: Used for iterating over a sequence (such as a list, tuple, string, or range) to perform a certain operation for each item in the sequence.

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
#while:

#Significance: Used for creating loops that execute a block of code as long as a specified condition is true.
count = 0
while count < 5:
    print(count)
    count += 1
    
#True and False:

#Significance: Boolean literals representing true and false values, commonly used in conditional expressions.
is_true = True
is_false = False

#def:

#Significance: Used to define a function in Python, allowing you to encapsulate a block of code with a name and parameters for reuse.
def greet(name):
    print("Hello, " + name + "!")

greet("Debashis")


```
